```python
from glob import glob
globing = glob('./data/*')
globing = [x.split('/')[2] for x in globing]
label2id, id2label = dict(), dict()
for i, label in enumerate(globing):
    label2id[label] = i
    id2label[i] = label
```

```python
import pandas as pd
from PIL import Image
import numpy as np
labels = []
images = []
for j in checking:
    label = label2id[j.split('/')[2]]
    path=  j
    labels.append(label)
    images.append(path)
```

```python
import pandas as pd
df = pd.DataFrame()
df['label']=labels
df['img']=images
df.to_csv('./dataset_for_classification.csv',index=False)
```
